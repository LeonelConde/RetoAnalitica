```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
dataframe = pd.read_csv("avocado.csv")
dataframe.head()
```

```python
dataframe.describe()
```

```python
print(dataframe.groupby('year').size())
```

```python
dataframe.drop(['year','Unnamed: 0','4046','4225','4770'],1).hist()
plt.show()
```

```python
sb.pairplot(dataframe.dropna(), hue='year',height=4,vars=["AveragePrice","Total Volume","Total Bags","Small Bags", "Large Bags","XLarge Bags"],kind='scatter')
```

```python
X = np.array(dataframe[["Total Volume","Total Bags","Small Bags", "Large Bags","XLarge Bags"]])
y = np.array(dataframe['year'])
X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green','blue','cyan','yellow','orange','black','pink','brown','purple']
asignar=[]
for row in y:
    asignar.append(colores[row-2015])
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
labels = kmeans.predict(X)
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000)
```

```python
copy =  pd.DataFrame()
copy['year']=dataframe['year'].values
copy['AveragePrice']=dataframe['AveragePrice'].values
copy['label'] = labels;
cantidadGrupo =  pd.DataFrame()
cantidadGrupo['color']=colores
cantidadGrupo['cantidad']=copy.groupby('label').size()
cantidadGrupo
```

```python
group_referrer_index = copy['label'] ==0
group_referrals = copy[group_referrer_index]

diversidadGrupo =  pd.DataFrame()
diversidadGrupo['year']=[0,1,2,3,4]
diversidadGrupo['cantidad']=group_referrals.groupby('year').size()
diversidadGrupo
```
